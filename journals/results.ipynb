```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Alle oorspronkelijke data die je had opgegeven
data = {
    'Method': ['Hybrid', 'Hybrid', 'Dense', 'Sparse', 'Sparse', 'Sparse'],
    'Type': ['hr', 'ict', 'ict', 'ict', 'ict', 'hr'],
    'avg_precision': [0.6083333333333333, 0.48166666666666674, 0.35500000000000004, 0.45, 0.75, 0.6333333333333333],
    'avg_recall': [1, 0.9333333333333333, 0.9333333333333333, 0.45, 0.75, 0.8],
    'avg_f1': [0.7066666666666668, 0.6083333333333333, 0.5021428571428572, 0.4333333333333333, 0.7222222222222222, 0.6833333333333333]
}

df = pd.DataFrame(data)

# Bereken de gemiddelde metrics per methode
average_metrics = df.groupby('Method').mean(numeric_only=True)

# Bepaal de gewenste volgorde voor de x-as
new_order = ['Dense', 'Sparse', 'Hybrid']
average_metrics = average_metrics.reindex(new_order)

# Genereer de staafdiagram
ax = average_metrics.plot(kind='bar', figsize=(12, 7), rot=0)

# Stel de titels en labels in
plt.title('Vergelijking van Gemiddelde Retrieval Metrics per Methode', fontsize=16)
plt.ylabel('Score', fontsize=12)
plt.xlabel('Methode', fontsize=12)
plt.ylim(0, 1)
plt.legend(title='Metrics', labels=['Precision', 'Recall', 'F1-score'])
plt.grid(axis='y', linestyle='--')

# Voeg de numerieke waarden boven de staven toe
for p in ax.patches:
    ax.annotate(f"{p.get_height():.2f}", 
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', 
                xytext=(0, 10), 
                textcoords='offset points')

# Zorg ervoor dat alle elementen netjes binnen de figuur passen
plt.tight_layout()

# Toon de grafiek
plt.show()
```

```python


import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Data voor de vergelijking tussen Hybrid en Hybrid met reranking
data = {
    'Method': ['Hybrid', 'Hybrid met reranking'],
    'avg_precision': [0.545000, 0.7619047619047619],
    'avg_recall': [0.966667, 0.964],
    'avg_f1': [0.657500, 0.8333333333333333]
}

# Maak een DataFrame aan van de data
df = pd.DataFrame(data)
df.set_index('Method', inplace=True)

# Genereer de staafdiagram
# rot=0 zorgt voor horizontale labels
ax = df.plot(kind='bar', figsize=(10, 7), rot=0)

# Stel de titels en labels in
plt.title('Vergelijking Hybrid vs. Hybrid met Reranking', fontsize=16)
plt.ylabel('Score', fontsize=12)
plt.xlabel('Methode', fontsize=12)
plt.ylim(0, 1.1)
plt.legend(title='Metrics', labels=['Precision', 'Recall', 'F1-score'])
plt.grid(axis='y', linestyle='--')

# Voeg de numerieke waarden boven de staven toe
for p in ax.patches:
    ax.annotate(f"{p.get_height():.2f}", 
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', 
                xytext=(0, 10), 
                textcoords='offset points')

# Zorg ervoor dat alle elementen netjes binnen de figuur passen
plt.tight_layout()

# Toon de grafiek
plt.show()
```
